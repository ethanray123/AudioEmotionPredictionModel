```python
from string import punctuation
from os import listdir
from collections import Counter
from nltk.corpus import stopwords
from keras.preprocessing.text import Tokenizer

import csv
import numpy as np

# load doc into memory
def load_doc(filename):
    # open the file as read only
    file = open(filename, 'r')
    # read all text
    text = file.read()
    # close the file
    file.close()
    return text


# turn a doc into clean tokens
def clean_doc(doc):
    # split into tokens by white space
    tokens = doc.split()
    # remove punctuation from each token
    table = str.maketrans('', '', punctuation)
    tokens = [w.translate(table) for w in tokens]
    # remove remaining tokens that are not alphabetic
    tokens = [word for word in tokens if word.isalpha()]
    # filter out stop words
    stop_words = set(stopwords.words('english'))
    tokens = [w for w in tokens if not w in stop_words]
    # filter out short tokens
    tokens = [word for word in tokens if len(word) > 1]
    return tokens


# load doc, clean and return line of tokens
def doc_to_line(filename):
    # load the doc
    doc = load_doc(filename)
    # clean doc
    tokens = clean_doc(doc)
    return ' '.join(tokens)


# load all docs in a directory and place data in a list
def read_docs(directory):
    lines = list()
    filenames = listdir(directory)
    filenames.sort()
    
#     print(filenames)
    # walk through all files in the folder
    for filename in filenames:
        # create the full path of the file to open
        path = directory + '/' + filename
        print(path)
        # load and clean the doc
        line = doc_to_line(path)
        # add to list
        lines.append(line)
    return lines


lines = read_docs('MER_bimodal_dataset/Lyrics_Dump')
wva = np.load('rated_words.npy')
print(wva[:,1])

print("# of Lyrics:",len(lines))
# print(lines[1])
max_words_per_line = len(lines[0].split())
for i in range(1,len(lines)):
    line = lines[i].split()
    if(len(line) > max_words_per_line):
        max_words_per_line = len(line)
print('maximum number of words per line: ',max_words_per_line)
```

```python
average_valence = []
average_arousal = []
for i in range(len(lines)): # len(lines)
    line = lines[i].split()
    sum_valence = 0
    sum_arousal = 0
    no_of_matched_words = 0
    for x in range(len(line)):
        if(line[x] in wva[:,0]):
            no_of_matched_words+=1
            sum_valence += float(wva[x][1])
            sum_arousal += float(wva[x][2])
            print("lyric",i+1,"word valence:",wva[x][1])
            print("lyric",i+1,"word arousal:",wva[x][2])
    print("average valence:", sum_valence/no_of_matched_words)
    print("average arousal:", sum_arousal/no_of_matched_words)
    average_valence.append(sum_valence/no_of_matched_words)
    average_arousal.append(sum_arousal/no_of_matched_words)
    if(len(line) > max_words_per_line):
        max_words_per_line = len(line)
print('maximum number of words per line: ',max_words_per_line)
```

```python
lyrics = []
for i in range(1,len(lines)):
    lyrics.append(lines[i].split())
    
filenames = listdir('MER_bimodal_dataset/Lyrics_Dump')
filenames.sort()

a = np.array([lyrics[0],average_valence[0],average_arousal[0]])
b = np.array([lyrics[1],average_valence[1],average_arousal[1]])
rated_lyrics = np.vstack((a,b))
print(rated_lyrics[1][2])
```
